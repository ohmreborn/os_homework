```python
import plotly.express as px
import pandas as pd
import numpy as np
```

```python
df_serial = pd.read_csv("result_serial.csv").dropna(axis=1)
display(df_serial)
fig = px.line(
    df_serial.T,
    title='time use in serial',
    labels={
        "index": "num thread",     # X-axis label
        "value": "time in seconds",   # Y-axis label
        "variable": "number for factorization"             # Legend title
    }
    )
fig.show()
```

```python
df_parallel = pd.read_csv("result_parallel.csv").dropna(axis=1)
display(df_parallel)
fig = px.line(
    df_parallel.T,
    title='time use in parallel',
    labels={
        "index": "num thread",     # X-axis label
        "value": "time in seconds",   # Y-axis label
        "variable": "number for factorization"             # Legend title
    }
    )
fig.show()
```

```python
df_total = df_serial + df_parallel
serial = df_serial / df_total
parallel = df_parallel / df_total
Amdahl = 1 / (serial + (parallel/ np.array(parallel.columns.astype(float))))
display(Amdahl)
fig = px.line(
    Amdahl.T,
    title="Amdahl's law",
    labels={
        "index": "num thread",     # X-axis label
        "value": "time in seconds",   # Y-axis label
        "variable": "number for factorization"             # Legend title
    }
    )
fig.show()
```
